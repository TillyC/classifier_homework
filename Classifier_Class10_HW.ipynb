```python
from stop_words import get_stop_words

# here's the template for how we want to structure our features
# {'first_keyword': slkdjfsl, 'second_keyword': lsdjfls, 'third_keyword': slkdjf}
def create_features(critique):
    # make all the words lowercase
    critique_description = critique.lower()
    # split the headline into a list of words
    words = critique_description.split()
    
   #while len(words) < 4:
        #words.append('None')
    #try adding some more keywords to improve accuracy (I added up to seven but this didn't affect it)
   # return {'first_keyword': words[0], 'second_keyword': words[1], 'third_keyword': words[2], 'fourth_keyword': words[3]}
    
    # using list comprehension to create a list of keywords
    #remove stopwords, full stops and numbers to aid accuracy
    keywords = [ x for x in words if x not in get_stop_words('english') and (".") not in x and not x.isdigit()] 
    while len(keywords) < 4:
        keywords.append('None')
    #try adding some more keywords to improve accuracy (I added up to seven but this didn't affect it)
    return {'first_keyword': keywords[0], 'second_keyword': keywords[1], 'third_keyword': keywords[2], 'fourth_keyword': keywords[3]}
```

```python
# bring together the rating(label), and the critique(features)
zipped_feature_and_labels = zip(critique, reviews.review_type)
```

```python
# we are creating a list of tuples
# we want () around our headline and category to show they are tuples
feature_sets = [ (create_features(x),y) for x,y in zipped_feature_and_labels   ]
feature_sets
```

```python
# always shuffle your featureset before classifying!
import random
random.shuffle(feature_sets)
```

```python
import math

split_num = math.floor(len(feature_sets) * .8)

# 80% of my data
training_set = feature_sets[:split_num]
# 20% of my data
testing_set = feature_sets[split_num:]
```

```python
# let's look at how many feature sets we have
len(feature_sets)
```

```python
import nltk
# create our classifier
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
# try out our classifier using a review
classifier.classify(create_features("With great effects (and a few surprises) it's a fun, Saturday-matinee adventure."))
```

```python
# determine the accuracy of our classifier
print(nltk.classify.accuracy(classifier, testing_set))
```
